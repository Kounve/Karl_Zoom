```python
#import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = pd.read_csv("data/zoominfo_title_data.csv")
df = pd.DataFrame(data)
df.head()
```

```python
# Change zoominfo titles column to uppercase
zoominfo = df["Zoominfo Title"].str.upper()
# Insert column into the DataFrame
df.insert(loc=10, column="Zoominfo Title Upper", value=zoominfo)
# Check to see if it worked!
df.head()
```

```python
actual = df["Actual Title"].str.upper()
df.insert(loc=11, column="Actual Title Upper", value=actual)
df.head()
```

```python
# Add underscore to column name so Python can query
df = df.rename(columns={'Vendor Title' : 'Vendor_Title', 'Zoominfo Title Upper' : 'Zoominfo_Title', 'Actual Title Upper' : 'Actual_Title'})
# Drop unwanted columns
clean = df.loc[:, ["Vendor_Title", 'Zoominfo_Title', 'Actual_Title']]
clean
```

```python
vendor_to_ actual = clean.query('Vendor_Title == Actual_Title')
```

```python
# Query where vendor title is the same as the actual title and where ZoomInfo is not the same
title_query = clean.query('Vendor_Title == Actual_Title & Zoominfo_Title != Actual_Title')
```

```python
zoom_to_actual = clean.query('Zoominfo_Title == Actual_Title')
```
